```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

plt.rcParams['figure.figsize'] = [15, 15]
```

```python
vidcap = cv2.VideoCapture('Basler acA2040-90umNIR (22821916)_20190221_200711434.mp4')
```

```python
success = True
count = 0
while(success):
    if count%10 == 0:
        success,image = vidcap.read()
        gray = image[500:1250,...,0]
        #gray = rgb2gray(image)
        cv2.imwrite("labeled/frame%d.jpg" % (count/10), gray)
        break
    count += 1
```

```python
ret3,th3 = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th3, cmap = plt.get_cmap('gray'))
plt.show()
```
